## counting log events

```python
events = ['loggedin', 'viewed', 'started' ,'submitted', 'uploaded', 'updated', 'searched', 'resumed','answered', 'attempted', 'abandoned']
eventdf = []

for user in users[0].head(5):
    eventls = {}
    eventls['id'] = user
    for event in enumerate(events):
        sql = %sql select count(*) from mdl_logstore_standard_log where userid = {user} and action = '{event}'
        eventls[event] = sql
        eventdf.append(eventls)
```
