# Class that has all functions to do all the tasks

```python
class Task():
    def __init__(self):
        self.d = None
    def db_connection(self):
        conn = %sql postgresql://postgres:password@localhost/moodle
        return conn
    def get_records_per_table(self,table):
        sql = %sql SELECT count(*)FROM {table}
        sql = print(sql)
        return sql
    def get_tables(self):
        query = %sql SELECT count(*) FROM information_schema.tables where TABLE_TYPE='BASE TABLE'
        return query
    def quiz_submission(self):
        query = %sql select count(id), EXTRACT(HOUR FROM to_timestamp(timecreated)) AS HOUR from mdl_logstore_standard_log where action='submitted' AND component='mod_quiz' group by EXTRACT            (HOUR FROM to_timestamp(timecreated))
        return query
    def events(self):
        query = %sql SELECT userid, COUNT(action) AS events FROM mdl_logstore_standard_log  WHERE action IN ('loggedin', 'viewed', 'started','submitted','uploaded', 'updated', 'searched' ,             'resumed','answered', 'attempted', 'abandoned') GROUP BY userid ORDER BY events
        return query
    def overall_grades(self):
        query = %sql SELECT userid, sum(finalgrade) FROM mdl_grade_grades GROUP BY userid ORDER BY userid
        return query
    def dedication_time(self):
        return None
    def login_count(self):
        query = %sql SELECT userid, COUNT(action) AS login_count \
                            FROM mdl_logstore_standard_log \
                            WHERE action = 'loggedin' \
                            GROUP BY userid \
                            ORDER BY login_count \
                            
        return query
    def activity_count(self):
        query = %sql SELECT userid, COUNT(action) AS activity_count \
                            FROM mdl_logstore_standard_log \
                            GROUP BY userid \
                            ORDER BY activity_count
        return query

    def percentages(self,count,perc):
        if count == "login":
            metric = self.login_count()
        elif count == "activity":
            metric = self.activity_count()
        else:
            metric = self.dedication_time()
        
        result = metric[: ((len(metric)*perc)//100)]

        return print(result)
```
